# lab05 - Logistic regression classifier

```python
import tensorflow as tf
import numpy as np
```

```python
x_data = [[1, 2],
          [2, 3],
          [3, 1],
          [4, 3],
          [5, 3],
          [6, 2]]
y_data = [[0],
          [0],
          [0],
          [1],
          [1],
          [1]]

tf.model = tf.keras.Sequential()
# use sigmoid activation for 0~1 problem
tf.model.add(tf.keras.layers.Dense(units=1, input_dim=2, activation='sigmoid'))

tf.model.compile(loss='mse', optimizer=tf.keras.optimizers.SGD(lr=0.01), metrics=['accuracy'])
tf.model.summary()

history = tf.model.fit(x_data, y_data, epochs=5000)

# Accuracy report
print("Accuracy: ", history.history['accuracy'][-1])
```

## diabetes

### loss='binary_crossentropy'

```python
xy = np.loadtxt('data-03-diabetes.csv', delimiter=',', dtype=np.float32)
x_data = xy[:, :-1]
y_data = xy[:, [-1]]

print(x_data.shape, y_data.shape)

tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(units=1, input_dim=x_data.shape[1], activation='sigmoid'))
tf.model.compile(loss='binary_crossentropy', optimizer=tf.keras.optimizers.SGD(lr=0.01), metrics=['accuracy'])
tf.model.summary()

history = tf.model.fit(x_data, y_data, epochs=500)

# accuracy
print("Accuracy: {}".format(history.history['accuracy'][-1]))

# predict a single data point
y_predict = tf.model.predict([[0.172837, 0.155254, 0, 0, 0, 0.0524376, -0.953478, -0.733333333]])
print("Predict: {}".format(y_predict))


# evaluating model
evaluate = tf.model.evaluate(x_data, y_data)
print("loss: {0}, accuracy: {1}".format(evaluate[0], evaluate[1]))
```
