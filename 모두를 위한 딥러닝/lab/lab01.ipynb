# Computational Graph

## case1) tf.Session() -> tf.compat.v1.Session()

### 1) Build graph(tensors) using TensorFlow operations

```python
node1 = tf.constant(3.0, tf.float32)
node2 = tf.constant(4.0)  # also tf.float32 implicitly
node3 = tf.add(node1, node2)
```

### 2) feed data and run graph (operation)

### 3) update variable in the graph (and return values)

```python
tf.compat.v1.disable_eager_execution()  # need to disable eager in TF2.x
sess = tf.compat.v1.Session()

result = sess.run([node1, node2])
result2 = sess.run(node3)

print(result)
print(result2)
```

## case2)

```python
node1 = tf.constant(3.0, tf.float32)
node2 = tf.constant(4.0)  # also tf.float32 implicitly

print("node1: ", node1)
print("node2: ", node2)
```

```python
@tf.function
def forward():
    return node1 + node2

out_a = forward()
print(out_a)
```

placeholder 해결 방법<br>
https://www.tensorflow.org/guide/variable?hl=ko

```python
W = tf.Variable(tf.ones(shape=(2,2)), name="W")
b = tf.Variable(tf.zeros(shape=(2)), name="b")

@tf.function
def forward(x):
  return W * x + b

out_a = forward([1,0])
print(out_a)
```

```python
@tf.function
def adder(a, b):
    return a + b

print(adder(W, b))
```
